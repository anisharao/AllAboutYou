```python

# Getting Datetime from timestamp
# date_time = datetime.fromtimestamp(1887639468)
# print("Datetime from timestamp:", date_time)


def submitact():
    Age_ip = int(Age.get())
    Gender_ip = Gender.get()
    Country_ip = Country.get()
    State_ip = State.get()
    Self_employed_ip = Self_employed.get()
    Family_history_ip = Family_history.get()
    Treatment_ip = Treatment.get()
    Work_interfere_ip = Work_interfere.get()
    No_of_emp_ip = int(No_of_emp.get())
    Remote_work_ip  = Remote_work.get()
    Tech_company_ip = Tech_company.get()
    Benefits_ip = Benefits.get()
    ct = datetime.now()
    Care_options_ip=Care_options.get()
    Wellness_prg_ip=Wellness_prg.get()
    Seek_help_ip=  Seek_help.get()
    Anonymity_ip=  Anonymity.get()
    Leave_ip= Leave.get()
    Mental_health_cons_ip =Mental_health_cons.get()
    Physical_health_cons_ip   =Physical_health_cons.get()
    Coworkers_ip  =Coworkers.get()
    Supervisor_ip = Supervisor.get()
    Mental_health_int_ip = Mental_health_int.get()
    Physical_health_int_ip = Physical_health_int.get()
    MentalvsPhysical_ip = MentalvsPhysical.get()
    Observed_cons_ip =  Observed_cons.get()
    Comments_ip = Comments.get() 
    
    
    insertformdetails(ct,Age_ip,Gender_ip,Country_ip,State_ip,Self_employed_ip,Family_history_ip,Treatment_ip,Work_interfere_ip,No_of_emp_ip,Remote_work_ip,Tech_company_ip,Benefits_ip,Care_options_ip,Wellness_prg_ip,Seek_help_ip,Anonymity_ip,Leave_ip,Mental_health_cons_ip,Physical_health_cons_ip, Coworkers_ip,Supervisor_ip, Mental_health_int_ip,Physical_health_int_ip,MentalvsPhysical_ip,Observed_cons_ip,Comments_ip)

    
def insertformdetails(ct,Age_ip,Gender_ip,Country_ip,State_ip,Self_employed_ip,Family_history_ip,Treatment_ip,Work_interfere_ip,No_of_emp_ip,Remote_work_ip,Tech_company_ip,Benefits_ip,Care_options_ip,Wellness_prg_ip,Seek_help_ip,Anonymity_ip,Leave_ip,Mental_health_cons_ip,Physical_health_cons_ip, Coworkers_ip,Supervisor_ip, Mental_health_int_ip,Physical_health_int_ip,MentalvsPhysical_ip,Observed_cons_ip,Comments_ip):
    result = survey_data_collection.insert_one({'Timestamp':ct,'Age':Age_ip,'Gender':Gender_ip,'Country':Country_ip,'State':State_ip,'self_employed':Self_employed_ip,'family_history':Family_history_ip,'treatment':Treatment_ip,'work_interfere':Work_interfere_ip,'remote_work':Remote_work_ip,'Tech_company':Tech_company_ip,'Benefits':Benefits_ip,'Timestamp':ct,'care_options':Care_options_ip,'wellness_program':Wellness_prg_ip,'seek_help':Seek_help_ip,'anonymity':Anonymity_ip,'leave':Leave_ip,'mental_health_consequence':Mental_health_cons_ip,'phys_health_consequence':Physical_health_cons_ip,'coworker':Coworkers_ip,'supervisor':Supervisor_ip,'mental_helath_interview': Mental_health_int_ip,'phys_helath_interview':Physical_health_int_ip,'mental_vs_physical':MentalvsPhysical_ip,'obs_consequence':Observed_cons_ip,'comments':Comments_ip})
    print(result)
```
